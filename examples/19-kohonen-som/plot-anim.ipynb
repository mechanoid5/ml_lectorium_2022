```python
from sklearn.datasets import make_blobs

n_samples = 500
n_clusters = 17 # количество кластеров

X,_ = make_blobs( 
        n_samples=n_samples,
        n_features=2,
        centers=n_clusters,  
        cluster_std=rng.uniform(size=n_clusters)+.7
    )
X.shape
```

```python
from matplotlib import pyplot as plt
from matplotlib.animation import ArtistAnimation
from IPython.display import HTML    
```

```python
from tqdm import tqdm
```

```python
frames = []

data_lim = np.vstack([X.min(axis=0)-.5,X.max(axis=0)+.5,]).T
# fig = plt.figure(figsize=(5,5))
# ax = plt.axes(xlim=data_lim[0], ylim=data_lim[1])

fig,ax = plt.subplots()
ax.set_xlim(data_lim[0])
ax.set_ylim(data_lim[1])

for i in tqdm(range(100)):
    mu = rng.normal(size=(3,2))
    frames.append( [
        ax.scatter(X[:,0],X[:,1],color='b',marker='.'),
        ax.scatter(mu[:,0],mu[:,1],color='r',marker='o') ,
    ])

# anim = ArtistAnimation(fig, frames, interval=50, blit=True, repeat_delay=1000)
HTML(ArtistAnimation(fig,frames,interval=150,).to_html5_video())    
```
