```python
plt.figure(figsize=(16, 10), dpi= 80)  
plt.title("USArrests Dendograms", fontsize=22)  
dend = shc.dendrogram(
    shc.linkage( df[['Murder', 'Assault', 'UrbanPop', 'Rape']], method='ward' ), 
    labels=df.State.values, 
    color_threshold=100 
)
plt.xticks(fontsize=12)
plt.show()
```
